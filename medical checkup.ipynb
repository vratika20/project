```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
medi = pd.read_csv(r"C:\Users\rajes\Downloads\medical checkup.csv")

# Attempt to split "Blood Pressure (s/d)" into numeric columns
medi[['Systolic_BP', 'Diastolic_BP']] = medi['Blood Pressure (s/d)'].str.split('/', expand=True)
medi['Systolic_BP'] = pd.to_numeric(medi['Systolic_BP'], errors='coerce')
medi['Diastolic_BP'] = pd.to_numeric(medi['Diastolic_BP'], errors='coerce')

# Group data by age and gender to calculate average metrics for each age group
grouped_medi = medi.groupby(['Age (years)', 'Gender']).mean(numeric_only=True).reset_index()

# Set up the figure size
plt.figure(figsize=(14, 10))

# Line chart for Cholesterol Levels by Age and Gender
plt.subplot(3, 1, 1)
sns.lineplot(data=grouped_medi, x='Age (years)', y='Cholesterol Level (mg/dL)', hue='Gender', marker='o')
plt.title("Cholesterol Levels by Age and Gender")
plt.xlabel("Age (years)")
plt.ylabel("Cholesterol Level (mg/dL)")

# Line chart for BMI by Age and Gender
plt.subplot(3, 1, 2)
sns.lineplot(data=grouped_medi, x='Age (years)', y='BMI', hue='Gender', marker='o')
plt.title("BMI by Age and Gender")
plt.xlabel("Age (years)")
plt.ylabel("BMI")

# Line chart for Systolic Blood Pressure by Age and Gender
plt.subplot(3, 1, 3)
sns.lineplot(data=grouped_medi, x='Age (years)', y='Systolic_BP', hue='Gender', marker='o')
plt.title("Systolic Blood Pressure by Age and Gender")
plt.xlabel("Age (years)")
plt.ylabel("Systolic BP (mmHg)")

# Adjust layout for clarity
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv(r"C:\Users\rajes\Downloads\medical checkup.csv")

# Convert categorical data into numeric values
data['Physical Activity Level'] = data['Physical Activity Level'].map({'Low': 1, 'Moderate': 2, 'High': 3})
data['Sleep Patterns'] = data['Sleep Patterns'].map({'Normal': 1, 'Insomnia': 2, 'Irregular': 3})
data['Mental Health Status'] = data['Mental Health Status'].map({'Poor': 1, 'Good': 2, 'Excellent': 3})

# Select relevant columns
selected_data = data[['Physical Activity Level', 'Sleep Patterns', 'Mental Health Status', 'Stress Levels']]

# Calculate the correlation matrix
correlation_matrix = selected_data.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Heatmap of Physical Activity, Sleep, Mental Health, and Stress Levels')
plt.show()
```

```python
5.	How do education and income levels correlate with key health outcomes such as mental health status, blood glucose levels, or cholesterol?
```

```python



mental_health_dist = data.groupby(['Education Level', 'Income Level', 'Mental Health Status']).size().unstack().fillna(0)

 Prepare labels for each education level
education_levels = mental_health_dist.index.get_level_values(0).unique()
income_levels = mental_health_dist.columns


colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0', '#ffb3e6']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle("Distribution of Mental Health Status by Education and Income Levels", fontsize=16)
axes = axes.flatten()

for i, edu_level in enumerate(education_levels):
    edu_data = mental_health_dist.loc[edu_level].sum(axis=1)
    
    axes[i].pie(edu_data.values, labels=edu_data.index, autopct='%1.1f%%', colors=colors, startangle=140)
    axes[i].set_title(f'Education Level: {edu_level}')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python


# Group data by Education Level and Income Level, then calculate average values for key health outcomes
health_summary = data.groupby(['Education Level', 'Income Level']).agg({
    'Blood Glucose Level (mg/dL)': 'mean',
    'Cholesterol Level (mg/dL)': 'mean'
}).reset_index()

# Set up the figure
plt.figure(figsize=(14, 6))

# Line chart for Blood Glucose Level by Education and Income Level
plt.subplot(1, 2, 1)
for income_level in health_summary['Income Level'].unique():
    subset = health_summary[health_summary['Income Level'] == income_level]
    plt.plot(subset['Education Level'], subset['Blood Glucose Level (mg/dL)'], marker='o', label=income_level)
plt.title('Average Blood Glucose Level by Education and Income Level')
plt.xlabel('Education Level')
plt.ylabel('Blood Glucose Level (mg/dL)')
plt.legend(title='Income Level')


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))
# Line chart for Cholesterol Level by Education and Income Level
plt.subplot(1, 2, 2)
for income_level in health_summary['Income Level'].unique():
    subset = health_summary[health_summary['Income Level'] == income_level]
    plt.plot(subset['Education Level'], subset['Cholesterol Level (mg/dL)'], marker='o', label=income_level)
plt.title('Average Cholesterol Level by Education and Income Level')
plt.xlabel('Education Level')
plt.ylabel('Cholesterol Level (mg/dL)')
plt.legend(title='Income Level')
plt.tight_layout()
plt.show()
```
